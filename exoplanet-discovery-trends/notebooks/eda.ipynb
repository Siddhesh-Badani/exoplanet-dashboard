```python
import sys
from pathlib import Path

ROOT = Path().resolve().parent
sys.path.insert(0, str(ROOT / 'src'))

import pandas as pd
import plotly.express as px
from fetch_data import CACHE_PATH
from process import load_and_process

data = load_and_process(CACHE_PATH)
df = data['df_raw']
print(f'{len(df):,} rows loaded')
```

```python
print('Shape:', df.shape)
print()
print('Dtypes:')
print(df.dtypes)
print()
print('Null counts:')
print(df.isnull().sum().sort_values(ascending=False))
```

```python
fig = px.histogram(
    df.dropna(subset=['disc_year']),
    x='disc_year',
    nbins=40,
    title='Distribution of Discovery Years',
    labels={'disc_year': 'Year', 'count': 'Planets'},
    template='plotly_white',
)
fig.show()
```

```python
method_counts = df['discoverymethod'].value_counts().reset_index()
method_counts.columns = ['method', 'count']

fig = px.bar(
    method_counts,
    x='count',
    y='method',
    orientation='h',
    title='Planet Count by Detection Method',
    labels={'method': '', 'count': 'Planets'},
    template='plotly_white',
    color='count',
    color_continuous_scale='Blues',
)
fig.show()
```
